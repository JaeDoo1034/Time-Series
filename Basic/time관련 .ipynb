```python
################## datetime vs time
# 시계열 데이터를 다룸에 있어서 사람에게 친숙한 형태는
# 서로 다른 언어, 서로 다른 시간대에 있는 데이터를 다루는데 timestamp방식이 효율적
import time # timestamp방식
from datetime import datetime,timedelta # 우리에게 익숙한 시간표현

timestamp_now = time.time() # timestamp방식 ~ 1970년 1월1일 0시 0분 1초부터 누적되는 시간~ 절대 중복이 일어나지 않은 시간.
print(timestamp_now)

datetime_now = datetime.now()
print(datetime_now)


delta_datetime = timedelta(days= -30)    # 30일 전으로 돌림
print(datetime_now + delta_datetime)


end_datetime = datetime_now + delta_datetime
print(end_datetime.timetuple())          # timetuple타입 ; 시간, 일자 별로 값이 떨어지게끔.
end_timestamp = time.mktime(end_datetime.timetuple())
# time.mktime : timestamp타입으로 바꾸어 주는 함수 ~>  datetime 형식 -> timestamp형식
print(end_timestamp)

# timestamp -> datetime바꾸자!
print(datetime.fromtimestamp(timestamp_now))
print(end_timestamp)
print(datetime.fromtimestamp(end_timestamp))

# timestamp값은 늘 새로운 값이다!!
# - 1970년 1월 1일 0시 0분 1초부터 누적되므로

```

```python
################## **시간 관련 된 모듈 임포트#####################
# ex) "2019-07-05" 하나의 문자열 ~ 이것으로 날짜 계산하기 힘들다 ~ 문자열을 datetime으로 만들어서 계산하자 (datetime끼리 계산가능)
# 목적 : 문자열 <-> datetime 변환 연습!
# datetime에서 사람마다 날짜 표현하는 방식이 중요하므로 포멧이 중요!!
# To 문자열 : str()  / To. datetime : datetime()

import datetime

# 시간 전체에 관련된 변수 설정
# 현재 년도, 월, 일, 시간이 저장된다.
now = datetime.datetime.now()
print('now : ',now)

# 현재 년도만 추출
print('년도 =', now.year)
print('월 =', now.month)
print('일 =', now.day)
print('시 =', now.hour)

#

# 퀴즈
# 1) datetime -> 문자열
print("오늘은 {}년 {}월 {}일이빈다.".format(now.year,now.month,now.day))
print("작성일자 {}".format(now.strftime('%Y-%m-%d %H:%M:%S'))) # 하나의 문자열로 간단히 표현
# strftime : datetime형식을 str로 리턴되고 포멧팅을 해주는 함수.

print('현재시간은 {}시 {}분입니다.'.format(now.hour,now.minute)) # now.hour : 24시간 기준으로 표현
print('현재시간은 {}시 {}분입니다.'.format(now.hour-12,now.minute)) # now.hour-12 : 12시간 기준으로 표현

# 2) 문자열 -> datetime
newdt = datetime.datetime.strptime('2019-07-05','%Y-%m-%d') # (입력받은 것, 포메팅하려는 형식)
# strfpime : 문자열형식을 datetime로 리턴되고 포메팅을 해주는 함수.
print(newdt,type(newdt))

# 정리
# strftime : datetime형식을 문자열로 리턴되고 포메팅을 해주는 함수 ~ 형식 : strftime('%Y-%m-%d %H:%M:%S)
# strfpime : 문자열형식을 datetime로 리턴되고 포메팅을 해주는 함수 ~ 형식 : strptime('문자열 입력','%Y-%m-%d')
# 연도 ~ Y : 4자리 표현 / y = 2자리표현   ,  월 : m  , 일 :  d
# 시간 ~ H : 24시 기준으로 표현 / h : 12시 기준으로 표현 ,  분 : M , 초 : S

# 입력
Y = input()
now = datetime.datetime.now()
bitrh = datetime.datetime.strptime(Y,'%Y-%m-%d')
Day_difference = now-bitrh
Month_difference = Day_difference/30



```
